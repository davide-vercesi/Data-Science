### 12. For each year, compute the total amount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st, 2016, planned expiration time January 30th 2018, and amount 5000USD has an amount of 5000USD * 31 / (31+365+30) = 363.85 for 2016, 5000USD * 365 / (31+365+30) = 4284.04 for 2017, and 5000USD * 30 / (31+365+30) = 352.11 for 2018.

```python
# la funzione prende le date di inizio e fine, e ritorna il totale dei giorni che intercorrono tra queste due date, dividendo per anno
def day_calc(date1, date2):
    (begin,end) = (date1,date2) if date1<date2 else (date2,date1) #controllo di verifica che le date nel dataframe siano in ordine
    date_list = list() #primo valore data di inizio, secondo valore 1 gennaio anno successivo... fino alla data di fine
    date_list.append(begin)
    for y in range(begin.year+1, end.year+1): #.year metodo datetime mi da l'anno del valore
        date_list.append(pd.Timestamp(year=y, month=1, day=1, tz='utc')) #aggiungo timestamp del 1 gennaio dell'anno con cui entro nel ciclo(anno inizio +1) e così via...
    date_list.append(end) #ho lista con valore inizio e ogni primo gennaio degli anni in mezzo e infine la data finale
    
    res = dict()
    for i in range(len(date_list)-1):
        res[date_list[i].year] = (date_list[i+1] - date_list[i]).days # faccio un dizionario con chiave l'anno e come valore il numero di giorno passati dalla chiave alla sua successiva che sono gli anni

    return pd.Series(res, name='DateValue') # con la series la chiave diventa indice e il valore rimane valore
```

```python
loans['planned_expiration_time'] = pd.to_datetime(loans['planned_expiration_time'])
loans['disburse_time'] = pd.to_datetime(loans['disburse_time'])
loans.head()
```

```python
loans_drop = loans.dropna() #abbiamo tenuto i NaN fino ad ad adesso ma  ora romperebbero la funzione, perciò li togliamo
loans_drop['duration'] = loans_drop['duration'].dt.days
days = loans_drop.apply(lambda x: day_calc(x['planned_expiration_time'],x['disburse_time']), axis = 1) # la x sarebbe loans, axis=1 perchè lo svolge per riga
days.head()
```

```python
days.fillna(0, inplace= True)
days = pd.concat([days, loans_drop['loan_id']
                 , loans_drop['loan_amount']
                 , loans_drop['duration'] ]
                , axis = 1, sort = True)

days.head()
```

```python
amounts = days[[2011,2012,2013,2014,2015,2016,2017,2018
                        ]].div(days.duration, axis=0).mul(days.loan_amount, axis = 0)

amounts = pd.concat([days['loan_id'], amounts],  axis = 1 ) # divide per duration così prende percentuale di giorni appartenenti a quell'anno poi moltiplica per loan amount così ha la percentuale di loan amount che appartiene a quell'anno 
amounts = amounts.melt(id_vars=["loan_id"], # .melt() compie una trasformazione unpivot sul dataframe
        var_name="Year", 
        value_name="amount_per_year")

amounts.groupby(['loan_id', 'Year']).sum()
```

### 17. Find the country with the highest rate of irregular repayment interval.

```python
repayments = pd.DataFrame(loans.groupby(['country_code', 'repayment_interval'], as_index=False)['loan_id'].count())
repayments = repayments.merge(repayments.groupby('country_code')['loan_id'].sum(), on='country_code')
repayments = repayments[repayments['repayment_interval'] == 'irregular']
repayments['irreg_ratio'] = repayments['loan_id_x'] / repayments['loan_id_y']
repayments.rename(columns={'loan_id_x': 'irregular_count', 'loan_id_y': 'total_count'}, inplace=True)
repayments.head()
```

```python
repayments.loc[repayments['irreg_ratio'].idxmax(),['country_code', 'irreg_ratio']] #Sri Lanka
```

### 18. Find the country with the highest fraction of loaned amount with irregular repayment interval.

```python
irreg_loans = loans[loans['repayment_interval'] == 'irregular'].groupby('country_code', as_index=False)[['country_code', 'loan_amount']].sum()
irreg_loans.rename(columns={'loan_amount': 'irreg_amount'}, inplace=True)
irreg_loans = irreg_loans.merge(loans.groupby('country_code', as_index=False)['loan_amount'].sum())
irreg_loans['irreg_ratio'] = irreg_loans['irreg_amount'] / irreg_loans['loan_amount']
irreg_loans.head()
```

```python
irreg_loans.loc[irreg_loans['irreg_ratio'].idxmax(),['country_code', 'irreg_ratio']] #Sri Lanka
```
