```python
from datetime import datetime
def convert_time(time_u): 
    ts = time_u

    # if you encounter a "year is out of range" error the timestamp
    # may be in milliseconds, try `ts /= 1000` in that case
    return datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
```

```python
df["create_time"] = df["create_time"].apply(lambda x:convert_time(x))
```

```python
df_auth= pd.DataFrame(combinations(author,2), columns=['auth1','auth2'])
```

```python
def music_com(row):
    
    auth1= row.auth1
    auth2 = row.auth2

    mus_auth1= list(df_fuzzy[df_fuzzy["author_name"]== auth1]["music_title"].unique())
    mus_auth2= list(df_fuzzy[df_fuzzy["author_name"]== auth2]["music_title"].unique())
    return len((set(mus_auth1).intersection(set(mus_auth2))))

    

    
```

```python
def music_songs(row):
    
    auth1= row.auth1
    auth2 = row.auth2

    mus_auth1= list(df_fuzzy[df_fuzzy["author_name"]== auth1]["music_title"].unique())
    mus_auth2= list(df_fuzzy[df_fuzzy["author_name"]== auth2]["music_title"].unique())
    return list((set(mus_auth1).intersection(set(mus_auth2))))
```

```python
df_auth["com"] = df_auth.apply(music_com, axis=1)
df_auth["songs"]= df_auth.apply(music_songs, axis=1)
```
